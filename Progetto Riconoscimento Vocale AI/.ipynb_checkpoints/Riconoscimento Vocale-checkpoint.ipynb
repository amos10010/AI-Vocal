```python
from scipy.io import wavfile
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
scaler = StandardScaler()
svc = SVC()
t=10
c=100
```

```python
fsA, amos = wavfile.read("Prova.wav")
fsL, lorenzo = wavfile.read("Prova1.wav")

fsA -= 10
fsL -= 10
```

```python
freqEq = fsA*t
```

```python
amos = amos[:freqEq]
lorenzo = lorenzo[:freqEq]
```

```python
audioA = pd.Series(amos)
audioL = pd.Series(lorenzo)
```

```python
audioA.describe()
```

```python
audioL.describe()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(np.arange(freqEq), audioA)
plt.show()

plt.figure(figsize=(15, 5))
plt.plot(np.arange(freqEq), audioL)
plt.show()
```

```python
batchAmos = np.zeros(int(freqEq/c))
batchLore = np.zeros(int(freqEq/c))
```

```python
for i in range(int(fsA/100)):
    f = i*100
    batchAmos[i] = amos[f:f+100]
for i in range(int(fsA/100)):
    f = i*100
    batchLore[i] = lorenzo[f:f+100]
```

```python
batchFull = np.zeros(int(freqEq/c))
target = np.zeros(int(freqEq/c))
```

```python
for i,x in enumerate(batchAmos):
    batchFull[i] = x
    target[i] = 0
for i, y in enumerate(batchLore):
    batchFull[i] = y
    target[i] = 1
```
